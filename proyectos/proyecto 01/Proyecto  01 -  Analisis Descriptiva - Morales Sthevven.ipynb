```python
# Medidas de tendencia central
media = df['VENT-UNIDAD'].mean()
mediana = df['VENT-UNIDAD'].median()
moda = df['VENT-UNIDAD'].mode().values[0]

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

```python
#GRAFICO SOBRE LA MARCA Y EL VENT-UNIDAD
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado 'df' que contiene las columnas 'MARCA' y 'VENT-UNIDAD'

# Agrupar las marcas en categorías más amplias
def agrupar_marcas(marca):
    if 'SHINERAY' in marca:
        return 'Grupo1'
    elif 'CHEVROLET' in marca:
        return 'Grupo2'
    elif 'KIA' in marca:
        return 'Grupo3'
    elif 'RANGER' in marca:
        return 'Grupo4'
    elif 'DAYTONA' in marca:
        return 'Grupo5'
    else:
        return 'Otros'

# Crear una nueva columna 'GRUPO' basada en la función de agrupación de marcas
df['GRUPO'] = df['MARCA'].apply(agrupar_marcas)

# para calcular la suma total de los valores de la columna 'VENT-UNIDAD' para cada grupo de marcas.
sum_ventas_grupo = df.groupby('GRUPO')['VENT-UNIDAD'].sum()

# Graficar las ventas por unidad por grupo utilizando un gráfico de barras
sum_ventas_grupo.plot(kind='bar')

# Configurar los ejes y el título del gráfico
plt.xlabel('GRUPO')
plt.ylabel('VENT-UNIDAD')
plt.title('Ventas por unidad por grupo de marcas')

# Mostrar el gráfico
plt.show()
```

```python
#se vera la relacion en la grafica entre la marca y el precio que este caso se le conoce como el avaluo
# Crear una nueva columna 'GRUPO' basada en la función de agrupación de marcas
#df['GRUPO'] = df['MARCA'].apply(agrupar_marcas)

#la media 
mean_avaluo_grupo = df.groupby('GRUPO')['AVALÚO'].mean()

# Graficar las ventas por unidad por grupo utilizando un gráfico de barras
mean_avaluo_grupo.plot(kind='bar')

# Configurar los ejes y el título del gráfico
plt.xlabel('GRUPO')
plt.ylabel('AVALUO')
plt.title('Precio por grupo de marcas')

# Mostrar el gráfico
plt.show()
```

```python
#Variacion del precio de cada auto a traves de la variable avaluo
# Create line chart of one column
plt.plot(df['AVALÚO'])
plt.title('Line Chart')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
# Graficar el gráfico de dispersión
# Extrae las columnas 'año y 'Grupo de Autos' del DataFrame
ano_modelo = df['AÑO MODELO']
grupo_marca = df['GRUPO']

# Graficar el gráfico de dispersión
plt.scatter(ano_modelo, grupo_marca)

# Configurar los ejes y el título del gráfico
plt.xlabel('AÑO MODELO ')
plt.ylabel('Grupo de Marcas')
plt.title('Gráfico de Dispersión - Año vs. Grupo de Marcas')

# Mostrar el gráfico
plt.show()
```

```python
#RELACION ENTRE TIPO COMBUSTIBLE CON LOS GRUPOS DE LAS MARCAS
# Agrupa los datos por el grupo de marcas y el tipo de combustibles
grupo_combustible = df.groupby(['GRUPO', 'TIPO COMBUSTIBLE']).size().unstack()

# Crea el gráfico de barras apiladas
grupo_combustible.plot(kind='bar', stacked=True,width=0.5)

# Configura los ejes y el título del gráfico
plt.xlabel('Grupo de Marcas')
plt.ylabel('Cantidad')
plt.title('Distribución de Tipo de Combustible por Grupo de Marcas')

# Muestra el gráfico
plt.show()
```

```python
# Agrupa los datos por la combinación de marca y modelo, y suma las ventas para obtener el total de ventas por cada combinación
ventas_totales = df.groupby(['MARCA', 'MODELO'])['VENT-UNIDAD'].sum().reset_index()

# Ordena los resultados en orden descendente según las ventas totales
ventas_totales = ventas_totales.sort_values('VENT-UNIDAD', ascending=False)

# Obtiene el auto y la marca más vendidos, que estarán en la primera fila del DataFrame
auto_mas_vendido = ventas_totales.iloc[0]['MODELO']
marca_mas_vendida = ventas_totales.iloc[0]['MARCA']

# Imprime los resultados
print(f"El modelo de auto  más vendido es: {auto_mas_vendido}")
print(f"La marca de auto más vendida es: {marca_mas_vendida}")
```

```python
#Lo que se hace es ver tendencias o patrones de los 10 marcas y modelos mas vendidos para asi poder tener una mejor idea sobre las preferencias de los usuarios
# Mostrar las primeras filas del DataFrame
print(df.head())

# Calcular estadísticas descriptivas básicas
print(df.describe())
#calcular las ventas totales
ventas_totales = df.groupby(['MARCA', 'MODELO'])['VENT-UNIDAD'].sum().reset_index()
#ordenar de forma descendente las ventas totales
ventas_totales = ventas_totales.sort_values('VENT-UNIDAD', ascending=False)
# Graficar las 10 marcas y modelos más vendidos
top_10 = ventas_totales.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_10['MARCA'] + ' ' + top_10['MODELO'], top_10['VENT-UNIDAD'])
plt.xlabel('Ventas')
plt.ylabel('Marca y Modelo')
plt.title('Top 10 de marcas y modelos más vendidos')
plt.show()
```

```python
# Calcular la cantidad de marcas por país de origen
marcas_por_pais = df['PAÍS'].value_counts()

# Crear un gráfico de barras para visualizar la cantidad de marcas por país
fig, ax = plt.subplots(figsize=(8, 6))
marcas_por_pais.plot(kind='bar', ax=ax)

# Personalizar el gráfico
ax.set_xlabel('País de Origen')
ax.set_ylabel('Cantidad de Marcas')
ax.set_title('Cantidad de Marcas por País de Origen')

# Mostrar el gráfico
plt.show()
```

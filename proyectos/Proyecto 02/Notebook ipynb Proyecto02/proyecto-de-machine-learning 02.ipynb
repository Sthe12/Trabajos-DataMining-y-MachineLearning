# Agregacion de la columna de clasificacion AQI

```python
# Obtener el valor máximo de las 5 columnas y asignarlo a la columna "AQI"
X_si['AQI'] = X_si[['SO2', 'NO2', 'PM10', 'CO', 'O3']].max(axis=1)
X_si
```

# Categorizar la columna AQI

```python
# Definir los límites y las etiquetas para cada rango
limits = [0, 50, 100, 150, 200, 250, 300, float('inf')]
labels = ['Excellent', 'Good', 'Sigthly Polluted','Lightly Polluted' , 'Moderately Polluted', 'Heavily Polluted', 'Severely Polluted']

# Asignar las etiquetas a los valores de la columna "AQI" basándose en los rangos definidos
X_si['AQI Category'] = pd.cut(X_si['AQI'], bins=limits, labels=labels, right=False)
X_si
```

# Creacion de modelos

## Modelo 1: Support Vector Classifier (SVC) 

### En este modelo, utilizamos un clasificador de vectores de soporte (SVC), utilizando un conjunto de datos dividido en conjuntos de entrenamiento y prueba. Luego, realiza predicciones en el conjunto de prueba y evalúa el rendimiento del modelo utilizando varias métricas de evaluación.

```python
from sklearn.model_selection import train_test_split
X = X_si.drop('AQI',axis=1)
X = X.drop('AQI Category',axis=1)
y = X_si['AQI Category']
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC

# Example for training and testing a Support Vector Classifier
model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

# Modelo 2: cross-validation 

### Se utiliza la técnica de validación cruzada o cross-validation para evaluar el rendimiento del modelo. La validación cruzada es una estrategia que divide el conjunto de datos en varias partes, llamadas "pliegues" o "folds", y luego entrena y evalúa el modelo en diferentes combinaciones de estos pliegues.

```python
from sklearn.model_selection import cross_val_score

# Example of cross-validation using 5 folds
scores = cross_val_score(model, X, y, cv=5)
print("Cross-Validation Scores:", scores)
print("Average Accuracy:", scores.mean())
```

# Modelo 3: Ajuste de hiperparámetros mediante Grid Search

### Involucra la técnica de ajuste de hiperparámetros utilizando la búsqueda en cuadrícula o Grid Search. Esta técnica se utiliza para encontrar la combinación óptima de hiperparámetros para un modelo de aprendizaje automático.

```python
from sklearn.model_selection import GridSearchCV

# Example of hyperparameter tuning using grid search
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print('Best parameters', best_params)
```
